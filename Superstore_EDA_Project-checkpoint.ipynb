# Superstore EDA Project
Exploratory Data Analysis on Sample Superstore Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Sample_Superstore_EDA_800_Rows.csv')
df.head()
```

```python
print(df.shape)
print(df.info())
df.describe()
```

```python
print(df.isnull().sum())
print('Duplicates:', df.duplicated().sum())
```

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title('Sales by Category')
plt.show()
```

```python
df.groupby('Region')['Profit'].sum().plot(kind='bar')
plt.title('Profit by Region')
plt.show()
```

```python
df['Order_Date']=pd.to_datetime(df['Order_Date'])
monthly=df.groupby(df['Order_Date'].dt.to_period('M'))['Sales'].sum()
monthly.plot(figsize=(10,5))
plt.show()
```

```python
sns.heatmap(df[['Sales','Quantity','Profit']].corr(),annot=True)
plt.show()
```
